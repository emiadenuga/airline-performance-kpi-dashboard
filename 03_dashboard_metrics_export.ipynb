## 3. KPI Definitions
The following definitions apply across all exported datasets:

- **Total Flights**
Count of scheduled flights

- **Cancellation Rate**
Cancelled flights / total scheduled flights

- **Delay Rate**
Flights arriving >15 minutes late / arrived flights

- **Severe Delay Rate**
Flights arriving >30 minutes late / arrived flights

- **Average Arrival Delay**
Mean arrival delay minutes across arrived flights

All rates are volume-weighted to reflect system-wide operational impact.

## 4. Tableau Dashboard Design Plan

### Dashboard Objective
This dashboard summarizes U.S. airline operational performance across the analyzed period. It is designed to help operations leadership quickly assess:
- Overall reliability
- Delay severity
- Seasonal performance trends
- Carrier & airport-level risk concentration

The dashboard supports data-driven decisions related to route optimization and performance monitoring.

----

### Dashboard Layout

#### Top Row - Executive KPI Snapshot

Displays the KPI cards for quick performance evaluation
- Total Flights 
- Cancellation Rate 
- Delay Rate 
- Severe Delay Rate
- Average Arrival Delay (minutes)
    
Purpose: 
 - Allow stakeholders to understand overall reliability within seconds

---

#### Middle Section - Monthly Performance Trends

Visuals:
- Delay Rate (line chart)
- Cancellation Rate (line chart)
- Average Arrival Delay (line chart)

Purpose:
- Identify seasonal patterns
- Detect performance improvements/decline
- Highlight disruption spikes during specific months

---

#### Bottom Left - Carrier Performance Comparison

Visual:
- Delay Rate by Carrier (sorted bar chart)
    
Purpose:
- Compare operational reliability across airlines
- Identify high and low performing carriers
- Support benchmarking and accountability discussions

---

#### Bottom Right - Geographic/Route Risk Analysis

Visual:
- Worst Routes and Origin Airports by Delay Rate (volume-filtered)
    
Purpose:
- Identify concentrated disruption areas
- Highlight high-impact operational bottlenecks
- Support targeted operational improvements

---

#### Intended Insights
This dashboard is designed to help answer the following questions:
- Which airlines have the highest delay and cancellation rates?
- Which airports or routes are most operationally risky?
- How has on-time performance changed over time?
- Are delays seasonal or systemic?
- Which carriers maintain consistent reliability?
- Where should leadership focus operational improvements first?

---

#### Connection to Business Value
By consolidating system-wide reliability metrics with carrier and geographic breakdowns, this dashboard enables:
- Proactive seasonal planning
- Targeted operational interventions
- Improved passenger experience
- Data-driven performance monitoring 

The dashboard translated analytical findings into actionable operational insight.
